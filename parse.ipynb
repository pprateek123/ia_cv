```python
def extract_information(text):
    # Define regex patterns
    # name_pattern = re.compile(r'\b(?:Mr\.|Ms\.|Mrs\.|Dr\.)\s*(\w+\s+\w+)\b', re.IGNORECASE)
    # email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    # phone_number_pattern = re.compile(r'\b(?:\+\d{1,2}\s?)?(\d{3}[.-]?\d{3}[.-]?\d{4})\b')
    name_pattern = re.compile(r'\b([A-Z][a-z]+ [A-Z][a-z]+)\b')
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    phone_number_pattern = re.compile(r'[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]')

    # Extract information
    name_match = name_pattern.search(text)
    email_match = email_pattern.search(text)
    phone_number_match = phone_number_pattern.findall(text)

    # Retrieve matched groups or return None if not found
    name = name_match.group(1) if name_match else None
    email = email_match.group() if email_match else None
    phone_number = phone_number_match if phone_number_match else None

    return name, email, phone_number
```

```python
def extract_text_from_docx(docx_path):
    txt = docx2txt.process(docx_path)
    if txt:
        return txt.replace('\t', ' ')
    return None
```
